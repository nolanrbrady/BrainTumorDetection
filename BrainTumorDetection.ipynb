```python
DATASET_DIR = "./BrainTumorsData/brain_tumor_dataset"

# image = imread(f'./10no.jpg', as_gray=True)
# imshow(image)
# image.shape, image

features = []
image_names = []

X_train = pd.DataFrame(columns = ['image_name','image_matrix', 'brain_tumor', 'path'])

# Populate X_train DataFrame with no tumor images
for images in os.listdir(f'{DATASET_DIR}/no'):
    # check if the image ends with png
    if (images.endswith(".jpg")):
        img_path = f'{DATASET_DIR}/no/{images}'
        image_data = imread(img_path, as_gray=True)
        no_tumor_list.append([images, image_data.flatten(), False, img_path])
        
# Populate X_train DataFrame with no tumor images
for images in os.listdir(f'{DATASET_DIR}/yes'):
    # check if the image ends with png
    if (images.endswith(".jpg")):
        img_path = f'{DATASET_DIR}/yes/{images}'
        image_data = imread(img_path, as_gray=True)
        tumor_list.append([images, image_data.flatten(), True, img_path])

training_list = no_tumor_list + tumor_list


X_train = pd.DataFrame(training_list, columns = ['image_name', 'image_matrix', 'tumor', 'path'])
print(X_train)
```

```python
# Function to Extract features from the images
def image_feature(direc):
    model = InceptionV3(weights='imagenet', include_top=False)
    features = [];
    img_name = [];
    for i in tqdm(os.listdir(direc)):
        
        img = image.load_img(i, target_size=(224,224))
        x = img_to_array(img)
        x = np.expand_dims(x,axis=0)
        x = preprocess_input(x)
        feat = model.predict(x)
        feat = feat.flatten()
        features.append(feat)
        img_name.append(i)
    return features,img_name
```

```python
direc = './BrainTumorsData/brain_tumor_dataset/combined'
img_features, img_name = image_feature(direc)
```

```python
# Creating Clusters
k = 2
clusters = KMeans(k, random_state = 0)
clusters.fit(img_features)
```

```python
image_cluster = pd.DataFrame(img_name, columns=['image'])
image_cluster["clusterid"] = clusters.labels_
image_cluster # 0 denotes cat and 1 denotes dog
```
